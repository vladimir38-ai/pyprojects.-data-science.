```python
# 6.2 Определите процент трудоустройства по гендерной принадлежности
```

```python
gender = data.pivot_table(index='find_new_job', columns='Пол', aggfunc={'count': 'sum'})
gender
```

```python
gender.columns = gender.columns.droplevel()
```

```python
perc_true_woman = 100 * gender.loc['True', 'Женский'] / (gender.loc['True', 'Женский'] + gender.loc['False', 'Женский'])
perc_true_woman
```

```python
perc_false_woman = 100 - perc_true_woman
perc_false_woman
```

```python
perc_true_man = 100 * gender.loc['True', 'Мужской'] / (gender.loc['True', 'Мужской'] + gender.loc['False', 'Мужской'])
perc_true_man
```

```python
perc_false_man = 100 - perc_true_man
perc_false_man
```

Создадим сводную таблицу по этим показателям, чтобы потом построить график

```python
dataset = pd.DataFrame({'Женщины': [perc_true_woman, perc_false_woman], 'Мужчины': [perc_true_man, perc_false_man]}, index=['Устроились на работу', 'Не устроились на работу'])
dataset
```

```python
dataset.plot(kind='bar', title='Трудоустройство по полу', ylabel='Процент')
plt.xticks(rotation=45)
plt.show()
```

Видим, что показатели в целом схожи. При этом мы помним, что женщин в принципе в датасете больше, чем мужчин.

```python
reg_list = data['region'].unique().tolist()
reg_list
```

```python
district_list = data['district'].unique().tolist()
district_list
```

```python
for district in district_list:
    print('ФЕДЕРАЛЬНЫЙ ОКРУГ:', district)
    d = data[data['district'] == district]
    reg = [] # Создадим списки, чтобы построить для каждого округа диаграмму.
    percent = []
    for region in reg_list:
        r = d[d['region'] == region]
        if len(r) > 0:
            cnt = r['count'].sum()
            employed = r[r['find_new_job'] == 'True']
            empl_cnt = employed['count'].sum()
            perc = 100 * empl_cnt / cnt 
            reg.append(region)
            percent.append(perc)
            print(f'{region}, (Процент трудящихся: {perc})')
            
    plt.figure(figsize=(12,8))
    plt.bar(reg, percent)
    plt.title(f'Процент трудоустроенных по регионам: {district}')
    plt.xlabel('Регион')
    plt.ylabel('Процент трудоустроенных (%)')
    plt.xticks(rotation=45)
    plt.show()    
           
        
```

Для цикла я создал два отдельных списка, состоящие из регионов и федеральных округов, которые затем использовал. Для каждого округа была построена диаграмма Федерального округа вместе с принадлежащими ему регионами. По графикам мы видим: хуже всего с трудоустройством дела обстоят в Северо-Кавказском Федеральном округе, где единственный регион, превышающий 10% трудоустройства - это Ставропольский край.
